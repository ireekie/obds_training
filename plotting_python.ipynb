#### Matplotlib

- Basic plotting package 
- Very customisable - every single line/pixel 
- Need alot of code to write a simple plot

```python
# Make x & y variables using numpy 
x = np.arange(0, 5, 0.1)
y = np.sin(x)

print('This is x:')
print(x)

print()

print('This is y:')
print(y)
```

### Matplot lib object orientated interface

```python

fig = plt.figure()   # create figure object - contain axes, graphics, text
ax = plt.axes()      # create axes - a single pane

ax.plot(x, y) # create continuous line plot

ax.set_xlim(0,10)
ax.set_ylim(-1.5,1.5)
ax.set_title('title of plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
plt.show()
```

### Matplot lib Pyplot interface

```python
plt.plot(x, y)
plt.xlim(0,10)
plt.ylim(-1.5,1.5)
plt.title('title of plt')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.show()
```

#### Scatterplot

```python
tips = sns.load_dataset("tips")
display(tips)

# set color to Yes/No depending on value in 'smoker column' 
sns.relplot(x="total_bill", 
            y="tip", 
            hue="smoker", 
            data=tips)

# as above plus shape of points for time column
sns.relplot(x="total_bill",
            y="tip",
            hue="smoker",
            style="time",
            data=tips);
```

```python

# col = Time -> create 2 plots (facets) split by time column 
sns.relplot(x="total_bill", 
            y="tip", 
            hue="smoker", 
            col="time", 
            data=tips)
```

#### Pairs plot

```python
iris = sns.load_dataset("iris")
display(iris.head())

# plot pairs plot
# assign plot to object 'g' - sets up grid
g = sns.PairGrid(iris, hue="species")

# plot histogram on diagonal
g.map_diag(plt.hist) 
# plot scatter plots in the offdiagonals
g.map_offdiag(plt.scatter)
# add legend to plot
g.add_legend();
```

### Lineplot example

```python
fmri = sns.load_dataset("fmri")
display(fmri.head())
sns.relplot(x="timepoint",
            y="signal", 
            hue="event", 
            kind="line", 
            data=fmri)

sns.relplot(x="timepoint", 
            y="signal", 
            hue="region", 
            style="event",
            dashes=False, 
            markers=True, 
            kind="line", 
            data=fmri);
```

### Facets

```python
sns.relplot(x="timepoint",
            y="signal",
            hue="subject",
            col="region",
            row="event",
            height=3,
            kind="line",
            estimator=None,
            data=fmri)
```

### Plot Categorical Data

```python
# create violin plots
g = sns.catplot(x="day", 
                y="total_bill", 
                kind="violin",
                inner=None,
                data=tips)

# add individual datapoints -> provide ax argument
sns.swarmplot(x="day",
              y="total_bill",
              color="k",
              size=3,
              data=tips,
              ax=g.ax) 
```

Identify gene with highest expression in each sample


```python
filtered_counts
```

```python
filtered_counts_geneindex = filtered_counts.set_index('Geneid')
maxValueIndexObj = filtered_counts_geneindex.idxmax()
maxValueIndexObj
```

Normalise read counts to total reads per sample & convert to 'counts per million (CPM)'


```python
sum_counts=filtered_counts_geneindex.sum() #sum columns - total counts of reads for each sample
display(sum_counts)
filtered_counts_geneindex.div(sum_counts) #divide by the sum of the columns
filtered_counts_CPM=filtered_counts_geneindex.div(sum_counts)*1e6
filtered_counts_CPM
```

Log transform normalised read counts


```python
filtered_counts_CPM_log=np.log2(filtered_counts_CPM +1)
filtered_counts_CPM_log
```

Create new dataframe of 100 genes with highest average variance in log cpm across conditions


```python
filtered_counts_CPM_log['variance']=filtered_counts_CPM_log.var(axis=1)
filtered_counts_CPM_log
```

```python
top_100_var=filtered_counts_CPM_log.sort_values(by=['variance'], ascending=False).head(100)
top_100_var
```

Plot the of expression of the top 5 most variable genes (log cpm) across all samples (line plot)


```python
top_5_var=top_100_var.head(5)
top_5_var_melt=top_5_var.drop('variance', axis=1).reset_index().melt(id_vars=['Geneid'],var_name='sample', value_name='counts') 
top_5_var_melt
```

```python
sns.lineplot(data=top_5_var_melt, x='sample', y='counts', hue='Geneid')
```

Plot a cluster map for top 100 most variable genes (log cpm) across all samples


```python
g = sns.clustermap(top_100_var.drop('variance', axis=1),figsize=(10,30))
```
